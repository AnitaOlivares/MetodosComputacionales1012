```python
# Importación de librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib

# Datos experimentales (en mm)
Ds2 = np.array([
    16.00, 16.15, 16.10, 15.85, 16.35, 16.20, 16.30, 15.95, 16.15,
    16.35, 16.35, 16.10, 16.30, 16.25, 16.20, 15.50, 16.50, 16.15,
    16.40, 16.50, 15.90, 15.65, 15.55, 16.60, 15.70, 16.85
])

# Cálculo de cantidad de intervalos usando la regla de Sturges
n = len(Ds2)
m = math.ceil(1 + 3.3 * math.log10(n))  # Redondeo hacia arriba

# Estadísticos
media = np.mean(Ds2)
desv_std = np.std(Ds2, ddof=1)  # Desviación estándar muestral

# Configuración visual
matplotlib.rcParams.update({'font.size': 16})
plt.figure(figsize=(12,6))

# Histograma con Sturges
R = max(Ds2)-min(Ds2)
bins = np.arange(min(Ds2), max(Ds2), R/m)
counts, bins_edges, _ = plt.hist(Ds2, bins=m, rwidth=0.9, color='silver', density=False, cumulative=False, align='mid')

# Líneas de media y desviación estándar
plt.vlines(media, 0, max(counts), colors='firebrick', ls='--', label=r'$\bar x$')
plt.hlines(max(counts)*0.6, media-desv_std, media+desv_std, colors='seagreen', ls='--', label=r'$\sigma$')

# Etiquetas y leyenda
plt.title('Histograma del diametro de los dados')
plt.xlabel('Diámetro / mm')
plt.ylabel('frecuencia')
plt.legend()
plt.grid(axis='y', alpha=0.55)
plt.show()

# Impresión de resultados
print(f"El diámetro promedio es {media:.2f} mm con una desviación estándar de {desv_std:.2f} mm.")
```

```python
# 📦 Importación de librerías
import numpy as np
import matplotlib.pyplot as plt
import math

# 🔁 Parámetros del experimento
np.random.seed(42)  # Para que los resultados sean reproducibles
repeticiones = 20
lanzamientos_por_serie = 9
p = 1/6  # Probabilidad de obtener un "4"

# 🎲 Simulación: lanzar el dado 9 veces, repetir eso 20 veces
resultados = [np.random.choice([1,2,3,4,5,6], size=lanzamientos_por_serie) for _ in range(repeticiones)]
frecuencias_de_4 = [np.count_nonzero(serie == 4) for serie in resultados]

# 📈 Cálculos teóricos y experimentales
n = lanzamientos_por_serie
media_teorica = n * p
desv_teorica = math.sqrt(n * p * (1 - p))
media_exp = np.mean(frecuencias_de_4)
desv_exp = np.std(frecuencias_de_4, ddof=1)

# 📊 Histograma de resultados
plt.figure(figsize=(10,6))
plt.hist(frecuencias_de_4, bins=5, align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.vlines(media_exp, 0, max(counts), colors='firebrick', ls='--', label=r'$\bar x$')
plt.hlines(max(counts)*0.6, media_exp-desv_exp, media_exp+desv_exp, colors='seagreen', ls='--', label=r'$\sigma$')
plt.xlabel("Veces que aparece el número 4 en 9 lanzamientos")
plt.ylabel("Frecuencia")
plt.title("Distribución experimental de obtener el número 4")
plt.grid(axis='y', alpha=0.6)
plt.show()



# 🧮 Probabilidad experimental de obtener un "4"
total_4s = sum(frecuencias_de_4)
total_lanzamientos = repeticiones * lanzamientos_por_serie
prob_4_exp = total_4s / total_lanzamientos * 100

# 🖨️ Resultados finales
print("📌 Resultados del experimento binomial:")
print(f"- Media teórica: {media_teorica:.2f}, experimental: {media_exp:.2f}")
print(f"- Desviación estándar teórica: {desv_teorica:.2f}, experimental: {desv_exp:.2f}")
print(f"- Probabilidad experimental de obtener un '4': {prob_4_exp:.2f}%")
```
